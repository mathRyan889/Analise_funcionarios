Agora queremos analisar a diferença salarial entre homens e mulheres


```python
dados.groupby('Sexo')[['Salário']].mean().sort_values('Salário')
```

```python
df_salarios_M_F = dados.groupby('Sexo')[['Salário']].mean().sort_values('Salário')
df_salarios_M_F.plot(kind= 'bar', figsize= (11,10), color = 'blue', title='Salário médio por Sexo')
```

Vamos analisar a média salarial apenas dos profissionais de exatas

```python
dados.Profissão.unique()
```

```python
profissões_exclusao = ['Médico','Enfermeiro','Vendedor','Advogado']
```

```python
dados.query('@profissões_exclusao not in Profissão')
```

```python
df = dados.query('@profissões_exclusao not in Profissão')
df.head()
```

```python
df_exatas_salarios = df.groupby('Profissão')[['Salário']].mean().sort_values('Salário')
df_exatas_salarios.plot(kind= 'bar', figsize= (11,10), color = 'blue', title='Salário médio por Profissão')
```

Vamos analisar quantos funcionarios ativos temos em cada profissão, e quais tem maior evasão

```python
df.groupby('Ativo')[['Profissão']].value_counts
```

```python
df_ativos = df.groupby('Ativo')[['Profissão']].value_counts()
df_ativos = df_ativos.loc['Sim']
df_ativos.plot(kind = 'barh', title='Quantidade de profissionais ativos', figsize=(14,10), color = 'blue')
```

```python
df_ativos = df.groupby('Ativo')[['Profissão']].value_counts()
df_ativos = df_ativos.loc['Não']
df_ativos.plot(kind = 'barh', title='Quantidade de profissionais ativos', figsize=(14,10), color = 'red')
```

# Onde estão localizados a maioria de nossos profissionais

```python
dados.groupby('Cidade')[['Profissão']].value_counts()
```

```python
df_localizacao = dados.groupby('Cidade')[['Profissão']].count()
df_localizacao.plot(kind = 'barh', title='Quantidade de profissionais por cidade', figsize=(14,10), color = 'green')
```

Como podemos ver a maioria dos nosso funcionarios residem em são de Salvador, para construimos instalações adequadas para esses funcionarios iremos analisar qual a maioria de funcionarios que residem nessa região

```python
df_localizacao = df.groupby('Cidade')[['Profissão',]].value_counts()
df_localizacao = df_localizacao.loc['Salvador']
df_localizacao.plot(kind = 'barh', title='Quantidade de profissionais por cidade', figsize=(14,10), color = 'green')
```

```python
df_localizacao = dados.groupby('Cidade')[['Profissão','Nome']].value_counts()
df_salvador = df_localizacao.loc['Salvador']
df_salvador.to_csv('Funcionarios_salvador.csv')
```

Vamos analisar quais cidades tem a maior faixa salarial

```python
salarios_por_cidade = dados.groupby('Cidade')[['Salário']].mean().sort_values('Salário')
salarios_por_cidade.plot(kind = 'barh', title='Quantidade de profissionais por cidade', figsize=(14,10), color = 'green')
```

```python
salarios_por_cidade.to_csv('Salarios_por_cidade.csv')
```
